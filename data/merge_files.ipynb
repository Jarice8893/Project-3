```python
import os, glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
path = "../Project-3/data"
all_files = glob.glob(os.path.join(path, "data_*.csv"))
display (all_files)
```

```python
for f in all_files:
    display(pd.read_csv(f,sep=','))
```

```python
all_csv = (pd.read_csv(f,sep=',') for f in all_files)
df_merged = pd.concat(all_csv, ignore_index = True)
df_merged.to_csv("merged.csv")
```

```python
df_merged.columns
```

```python
df_merged_aapl = df_merged.loc[df_merged["Name"].isin(["aapl"])]
df_merged_AR = df_merged.loc[df_merged["Name"].isin(["AR"])]
df_merged_msft = df_merged.loc[df_merged["Name"].isin(["msft"])]
df_merged_SPY = df_merged.loc[df_merged["Name"].isin(["SPY"])]
df_merged_TSLA = df_merged.loc[df_merged["Name"].isin(["TSLA"])]
df_merged_USO = df_merged.loc[df_merged["Name"].isin(["USO"])]
df_merged_aapl
```

```python
# number of trading days
df_merged_aapl.shape
```

```python
plt.figure(figsize =(16,8))
plt.title('AAPL Stock')
plt.xlabel('Days')
plt.ylabel('close price USD($)')
plt.plot(df_merged_aapl['Close'])
plt.show()
```

```python
#get close price 
df_merged_aapl = df_merged_aapl[['Close']]
```

```python
df_merged_aapl.head(4)
```

```python
future_days = 30

df_merged_aapl['Prediction'] = df_merged_aapl[['Close']].shift(-future_days)
```

```python
df_merged_aapl
```

```python
#feature data set (x) rows/days
x_aapl= np.array(df_merged_aapl.drop(['Prediction'],1))[:-future_days]
```

```python
print(x_aapl)
```

```python
#target data set (y)
y_aapl= np.array(df_merged_aapl['Prediction'])[:-future_days]
```

```python
print(y_aapl)
```

```python
#split the data into 75% training and 25% testing 
x_train_aapl, x_test, y_train_aapl, y_test = train_test_split(x_aapl, y_aapl, test_size=0.25)
```

```python
#create the models
#create the decision tree regressor model
tree = DecisionTreeRegressor().fit(x_train_aapl, y_train_aapl)
```

```python
#linear regression model
lr = LinearRegression().fit(x_train_aapl, y_train_aapl)
```

```python
x_future_aapl = df_merged_aapl.drop(['Prediction'], 1)[:-future_days]
```

```python
x_future_aapl = x_future_aapl.tail(future_days)
```

```python
x_future_aapl = np.array(x_future_aapl)
x_future_aapl
```

```python
#show the model tree prediction
tree_prediction = tree.predict(x_future_aapl)
print(tree_prediction)
```

```python
#show model linear regression predciton
lr_prediction = lr.predict(x_future_aapl)
```

```python
print(lr_prediction)
```

```python
#visualizing the data
predictions_aapl = tree_prediction
valid = df_merged_aapl[x_aapl.shape[0]:]
```

```python


valid['Prediction'] = predictions_aapl
plt.figure(figsize=(38,18))
plt.title('Model', fontsize = 17, weight = 'bold')
plt.xlabel('Days', fontsize = 17, weight = 'bold')
plt.ylabel('Close Price USD($)', fontsize = 17, weight = 'bold')
plt.plot(df_merged_aapl['Close'])
plt.plot(valid[['Close','Prediction']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```

```python

predictions_aapl = lr_prediction
valid = df_merged_aapl[x_aapl.shape[0]:]

valid['Prediction'] = predictions_aapl
plt.figure(figsize=(38,18))
plt.title('Model', fontsize = 17, weight = 'bold')
plt.xlabel('Days', fontsize = 17, weight = 'bold')
plt.ylabel('Close Price USD($)', fontsize = 17, weight = 'bold')
plt.plot(df_merged_aapl['Close'])
plt.plot(valid[['Close','Prediction']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```
